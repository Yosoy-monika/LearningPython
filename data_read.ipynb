```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import pandas as pd
```

```python
df= pd.read_csv('data.csv')
df
```

```python
plt.style.use('classic')
%matplotlib inline
mu = np.mean(df['1.41014'] )   
sigma = np.std(df['1.41014']) 

plt.figure(figsize=(15,15))
n,bins,patches=plt.hist(df['1.41014'],bins=18,density=True)
plt.legend(["normal fit"],loc='upper right',prop={"size":21})
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5* (1 / sigma * (bins - mu))**2))
plt.plot(bins,y, '--', c='r', linewidth =3)

plt.title("Histogram of Given Data")
plt.xlabel("Data points")
plt.ylabel("Height")
```
